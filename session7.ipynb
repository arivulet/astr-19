```python
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#create numpy array with 100 steps, n as int to hold how many steps
n = 100

x = np.linspace(0,1,n)
```

```python
#define our main and our exp function inside of that

def main():
    #def function that returns exp(x)
    def expfunc(x):
        return np.exp(x)

    #def y outside of the function so we can call the function
    y = expfunc(x)

    plt.plot(x,y)

    plt.xlabel("Time[Milliseconds]")
    
    plt.ylabel('Awesomeness')

    # show the plot!
    plt.savefig('exp_plot.pdf', format='pdf', bbox_inches='tight')

    plt.show()

if __name__ == "__main__":
    main()
```
